```python

import numpy as np
import matplotlib.pyplot as plt

def simular_caminata(pasos=100):
    # Definir posibles movimientos: +1 (derecha) o -1 (izquierda)
    movimientos = [1, -1]

    # Generar pasos aleatorios
    pasos_aleatorios = np.random.choice(movimientos, size=pasos)

    # Calcular la posicin acumulada (iniciando en 0)
    posiciones = np.cumsum(pasos_aleatorios)
    # Agregar el punto inicial
    posiciones = np.insert(posiciones, 0, 0)

    return posiciones

# Configuracin
n_pasos = 1000
camino = simular_caminata(n_pasos)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(camino)
plt.title(f"Caminata Aleatoria en 1D ({n_pasos} pasos)")
plt.xlabel("Pasos")
plt.ylabel("Posicin")
plt.grid(True)
plt.axhline(0, color='red', linestyle='--') # Lnea de inicio
plt.show()
```
